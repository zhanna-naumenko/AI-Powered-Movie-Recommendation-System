For this hackathon, I chose the TMDB + IMDB Merged Movies Dataset from Kaggle. This dataset combines movie information from The Movie Database (TMDb) and IMDb, providing a comprehensive collection of movie-related data.

It includes a wide range of attributes such as titles, release dates, ratings, genres, production companies, and more, making it a valuable resource for movie analysis, recommendation systems, and other data science applications.

However, before loading the dataset into MySQL, it needed to be cleaned and prepared.

Since not all columns are required for the movie database, we selected only the essential features needed for generating recommendations, such as genre, ratings, and titles, and created a refined dataset accordingly.

```python
keep_cols = [
    "id",
    "title",
    "genres",
    "overview",
    "keywords",
    "cast",
    "directors",
    "vote_average",
    "popularity",
    "release_date"
]
```

```python
new_movies = movies[keep_cols]
```

```python
new_movies.head()
```

```python
new_movies.isnull().sum()
```

Our most basic column is genre, so we can only delete missing values.

```python
new_movies = new_movies.dropna(subset=["title", "genres"])
```

```python
text_cols = ["overview", "keywords", "cast", "directors"]

for col in text_cols:
    new_movies[col] = new_movies[col].fillna("Unknown")
```

```python
new_movies["vote_average"] = new_movies["vote_average"].fillna(new_movies["vote_average"].median())
new_movies["popularity"] = new_movies["popularity"].fillna(new_movies["popularity"].median())
```

```python
new_movies["release_date"] = pd.to_datetime(new_movies["release_date"], errors="coerce")

new_movies["year"] = new_movies["release_date"].dt.year
new_movies["year"] = new_movies["year"].fillna(new_movies["year"].median())
```

```python
#Remove all extra quotes from all string columns
str_cols = new_movies.select_dtypes(include='object').columns
for col in str_cols:
    new_movies[col] = new_movies[col].str.replace('"', '', regex=False)
```

```python
new_movies["content"] = (
    new_movies["genres"] + " " +
    new_movies["keywords"] + " " +
    new_movies["overview"] + " " +
    new_movies["cast"] + " " +
    new_movies["directors"])
```

```python
print("Duplicate rows:", new_movies.duplicated().sum())
```

```python
new_movies.isnull().sum()
```

```python
new_movies["release_date"] = new_movies["release_date"].fillna(
    new_movies["year"].astype(int).astype(str) + "-01-01")
```

```python
new_movies.isnull().sum()
```

```python
#Fix the year column: remove .0 and convert to int
new_movies['year'] = new_movies['year'].astype(float).astype(int)
```

```python
cols_to_quote = [col for col in new_movies.columns if col != 'id']
for col in cols_to_quote:
    new_movies[col] = new_movies[col].apply(lambda x: f'"{x}"')

print(new_movies)
```

Now that our database is clean and without duplicates we can save it in .csv format and upload it to MySQL.

```python
new_movies.to_csv(
    "movies_cleaned.csv",
    index=False,
    encoding="utf-8")
```
